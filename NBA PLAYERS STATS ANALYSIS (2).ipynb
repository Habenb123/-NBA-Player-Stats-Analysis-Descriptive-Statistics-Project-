## DATA CLEANING

# MISSING VALUE HANDLING

```python
neme.isnull().sum()
```

160 Total Missing Values

```python

neme['Team'] = neme['Team'].astype('category')
neme['Pos'] = neme['Pos'].astype('category')


for col in ['Team', 'Pos']:
    if neme[col].dtype.name == 'category':
        if 'Unknown' not in neme[col].cat.categories:
            neme[col] = neme[col].cat.add_categories(['Unknown'])
    neme[col] = neme[col].fillna('Unknown')

zero_fill = ['Rk', 'G', 'GS', 'MP', '2P', '2PA', '3P', '3PA', 'BLK', 'STL', 'TOV',
             'PF', 'PTS', 'AST', 'REB', 'TRB', 'ORB', 'DRB', 'FTA', 'FT', 'FG', 'FGA', 'Trp-Dbl', 'Awards']

for col in zero_fill:
    if col in neme.columns:
        neme[col] = neme[col].fillna(0)

if 'Age' in neme.columns:
    neme['Age'] = neme['Age'].fillna(neme['Age'].median())

if '2P%' in neme.columns:
    neme['2P%'] = neme['2P%'].fillna(neme['2P%'].mean() if neme['2PA'].sum() > 0 else 0)

if 'eFG%' in neme.columns:
    neme['eFG%'] = neme['eFG%'].fillna(neme['eFG%'].mean() if neme['FGA'].sum() > 0 else 0)

if 'FG%' in neme.columns:
    neme['FG%'] = neme['FG%'].fillna(neme['FG%'].mean() if neme['FGA'].sum() > 0 else 0)

if '3P%' in neme.columns:
    neme['3P%'] = neme['3P%'].fillna(neme['3P%'].mean() if neme['3PA'].sum() > 0 else 0)

if 'FT%' in neme.columns:
    neme['FT%'] = neme['FT%'].fillna(neme['FT%'].mean() if neme['FTA'].sum() > 0 else 0)

if 'PER' in neme.columns:
    neme['PER'] = neme['PER'].fillna(neme['PER'].mean())

neme.dropna(subset=['Player'], inplace=True)

print("Missing values after cleaning:\n", neme.isnull().sum())
```

# Feature Engineering

## Efficiency Rating

```python
neme['Efficiency'] = (neme['PTS'] + neme['TRB'] + neme['AST'] + neme['STL'] + neme['BLK']) - (neme['FGA'] - neme['FG']) - (neme['FTA'] - neme['FT']) - neme['TOV']
```

## Scoring Per Minute

```python
neme['PTS_per_min'] = neme['PTS'] / neme['MP']
neme['PTS_per_min'].replace([np.inf, -np.inf], 0, inplace=True)
```

## Assist-to-Turnover Ratio

```python
neme['AST_TOV_ratio'] = neme['AST'] / neme['TOV']
neme['AST_TOV_ratio'].replace([np.inf, -np.inf], 0, inplace=True)
```

## Rebound Rate

```python
neme['REB_per_min'] = neme['TRB'] / neme['MP']
neme['REB_per_min'].replace([np.inf, -np.inf], 0, inplace=True)
```

## True Shooting Percentage (TS%)

```python
neme['TS%'] = neme['PTS'] / (2 * (neme['FGA'] + 0.44 * neme['FTA']))
neme['TS%'].replace([np.inf, -np.inf], 0, inplace=True)
```

## Radar Chart - Individual Player Performance

```python
from math import pi
import matplotlib.pyplot as plt

selected_player = 'LeBron James'  

stats = ['PTS', 'AST', 'TRB', 'STL', 'BLK']

player_data = neme[neme['Player'] == selected_player].iloc[0]

max_values = neme[stats].max()
normalized_stats = player_data[stats] / max_values
values = normalized_stats.tolist()
values += values[:1]  
labels = stats
angles = [n / float(len(labels)) * 2 * pi for n in range(len(labels))]
angles += angles[:1]

plt.figure(figsize=(8, 8))
plt.polar(angles, values, label=selected_player)
plt.xticks(angles[:-1], labels)
plt.title(f'Radar Chart - {selected_player} Performance')
plt.legend()
plt.show()
```

AST (Assists) and PTS (Points) are his strongest categories, both close to the maximum values among all players.

STL (Steals) is also relatively strong, though slightly lower.

TRB (Total Rebounds) is moderate—he's solid but not among the top in rebounding.

BLK (Blocks) is his weakest area in this comparison.

Overall, LeBron displays a balanced performance with dominance in scoring and playmaking. His defensive impact via blocks is lower than peers, but he compensates with steals and versatility.

# Clustering NBA Players Using KMeans

```python
# Step 1: Feature Selection

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

features = ['PTS', 'AST', 'TRB', 'STL', 'BLK', 'TOV', 'FG%', 'FT%', '3P%']
cluster_data = neme[features].copy()

# Fill missing values with column mean (safe measure)
cluster_data.fillna(cluster_data.mean(), inplace=True)

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)
```

```python
# Step 2: Find Optimal Number of Clusters (Elbow Method)

# Elbow method
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
# Step 3: Apply KMeans Clustering

# Let's assume k = 4 based on elbow method
kmeans = KMeans(n_clusters=4, random_state=42)
neme['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
# Step 4: Visualize Clusters

plt.figure(figsize=(8, 5))
sns.scatterplot(x=neme['PTS'], y=neme['AST'], hue=neme['Cluster'], palette='viridis')
plt.title('KMeans Clusters: Points vs Assists')
plt.xlabel('Points')
plt.ylabel('Assists')
plt.legend(title='Cluster')
plt.show()
```

This scatter plot shows KMeans clustering of players based on points and assists, forming four distinct groups. Cluster 0 (purple) contains high scorers and playmakers, while clusters 1–3 represent players with gradually decreasing contributions in both metrics.

```python
neme.groupby('Cluster')[features].mean().style.background_gradient(cmap='coolwarm')
```

The KMeans clustering successfully groups players into performance-based clusters (e.g., high scorers, defenders, role players), which can be useful for player comparison, scouting, or salary valuation.

Overall, this analysis provides a solid foundation for future, more advanced explorations like predictive modeling or team-level analysis.
